## **Handeling Categorical Variables**

```python
# As we see here the above data is not in numerical format so need to convert the X data to the numerical format
# for that we simply need to use the labelencoder function form the sklearn library


from sklearn import preprocessing
# we knoe that the Has Table booking and Has Online delivery column need to transform first

labelencoder = preprocessing.LabelEncoder()

X["Has Online delivery"] = labelencoder.fit_transform(X["Has Online delivery"])
```

```python
# for tge has table booking needs t be transform
X["Has Table booking"] = labelencoder.fit_transform(X["Has Table booking"])
```

```python
X
```

```python
y
```

```python
# has table 1 = YES and 0 = No
# has online 1 = YES and 0 = NO
```

## **Data SCaling**

```python
# AS we are using the liner regresstion model so theat we need to standard scalae the featues data for
# improving the model algorithm

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X = scaler.fit_transform(X)
X
```

## **Splitting The Dataset**

```python
# now splitting the data into the trainig and testing data as per 20% for the testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.2)
```

## **MOdel Training and Evaluation**

```python
#definging a function for the model result error check like MSE and MAE

from sklearn.metrics import mean_squared_error, median_absolute_error

def model_result(prediction):
    print("Mean sqared error:{}",format(mean_squared_error(y_test,prediction)))
    print("Median absolute error:{}",format(median_absolute_error(y_test, prediction)))
```

```python
#Now Creating a linear regression model for pedictions

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)

predictlr = lr.predict(X_test)

model_result(predictlr)
```

```python
# now creating a DecisionTree model for pedictions
# also using the GridSearchCV model for the decision tree model
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {
    "max_depth": [10,30],
    "min_samples_leaf": [1,2],
    "min_samples_split": [2,5]
}

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)
predictdt = dt.predict(X_test)


grid_tree = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5)
grid_tree.fit(X_train, y_train)
predictgr = grid_tree.predict(X_test)

model_result(predictdt)
model_result(predictgr)

```
